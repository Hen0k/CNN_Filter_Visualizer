```python

class Deconv(nn.Module):
    def __init__(self): # We introduce the __init__() function that will define the architecture of the generator.
        super(Deconv, self).__init__() # We inherit from the nn.Module tools.
        self.net = nn.Sequential(OrderedDict({
            "0" : nn.MaxUnpool2d(kernel_size=2, stride=2, padding=0),
            "1" : nn.ReLU(inplace=True),
            "2" : nn.ConvTranspose2d(512, 512, 3, stride=1, padding=1, output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros'),
            "3" : nn.ReLU(inplace=True),
            "4" : nn.ConvTranspose2d(512, 512, 3, stride=1, padding=1, output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros'),
            "5" : nn.ReLU(inplace=True),
            "6" : nn.ConvTranspose2d(512, 512, 3, stride=1, padding=1, output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros'),
            "7" : nn.MaxUnpool2d(kernel_size=2, stride=2, padding=0),
            "8" : nn.ReLU(inplace=True),
            "9" : nn.ConvTranspose2d(512, 512, 3, stride=1, padding=1, output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros'),
            "10" : nn.ReLU(inplace=True),
            "11" : nn.ConvTranspose2d(512, 512, 3, stride=1, padding=1, output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros'),
            "12" : nn.ReLU(inplace=True),
            "13" : nn.ConvTranspose2d(512, 256, 3, stride=1, padding=1, output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros'),
            "14" : nn.MaxUnpool2d(kernel_size=2, stride=2, padding=0),
            "15" : nn.ReLU(inplace=True),
            "16" : nn.ConvTranspose2d(256, 256, 3, stride=1, padding=1, output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros'),
            "17" : nn.ReLU(inplace=True),
            "18" : nn.ConvTranspose2d(256, 256, 3, stride=1, padding=1, output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros'),
            "19" : nn.ReLU(inplace=True),
            "20" : nn.ConvTranspose2d(256, 128, 3, stride=1, padding=1, output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros'),
            "21" : nn.MaxUnpool2d(kernel_size=2, stride=2, padding=0),
            "22" : nn.ReLU(inplace=True),
            "23" : nn.ConvTranspose2d(128, 128, 3, stride=1, padding=1, output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros'),
            "24" : nn.ReLU(inplace=True),
            "25" : nn.ConvTranspose2d(128, 64, 3, stride=1, padding=1, output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros'),
            "26" : nn.MaxUnpool2d(kernel_size=2, stride=2, padding=0),
            "27" : nn.ReLU(inplace=True),
            "28" : nn.ConvTranspose2d(64, 64, 3, stride=1, padding=1, output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros'),
            "29" : nn.ReLU(inplace=True),
            "30" : nn.ConvTranspose2d(64, 3, 3, stride=1, padding=1, output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros')

        }))
        
        self.outSize = [torch.Size([1, 64, 224, 224]),
                        torch.Size([1, 128, 112, 112]),
                        torch.Size([1, 256, 56, 56]),
                        torch.Size([1, 512, 28, 28]),
                        torch.Size([1, 512, 14, 14]),
                       ]
        
    def forward(self, x, layer, indices):
        j = -1
        k = len(indices) - 1
        
        print ("Input shape : ", str(x.shape))
        for i in range(31-layer, 31):
#             print (i)
            if isinstance(self.net[i], nn.MaxUnpool2d):
                x = self.net[i](x, indices[j], output_size=self.outSize[k])
                print ("back pool : ", x.shape)
                j -= 1
                k -= 1
            else:
#                 print (type(x))
                x = self.net[i](x)
                print("back : ", x.shape)
    
#                 print(self.net[i])
            
        return x # We return the output containing the generated images.
```

```python

class VGG16(nn.Module):
    def __init__(self):
        super(VGG16, self).__init__()
        # conv layers: (in_channel size, out_channels size, kernel_size, stride, padding)
        self.features = nn.Sequential(OrderedDict({
            str(0): nn.Conv2d(3, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            str(1): nn.ReLU(inplace=True),
            str(2): nn.Conv2d(64, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            str(3): nn.ReLU(inplace=True),
            str(4): nn.MaxPool2d(kernel_size=2, stride=2, padding=0, return_indices=True, dilation=1, ceil_mode=False),
            str(5): nn.Conv2d(64, 128, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            str(6): nn.ReLU(inplace=True),
            str(7): nn.Conv2d(128, 128, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            str(8): nn.ReLU(inplace=True),
            str(9): nn.MaxPool2d(kernel_size=2, stride=2, padding=0, return_indices=True, dilation=1, ceil_mode=False),
            str(10): nn.Conv2d(128, 256, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            str(11): nn.ReLU(inplace=True),
            str(12): nn.Conv2d(256, 256, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            str(13): nn.ReLU(inplace=True),
            str(14): nn.Conv2d(256, 256, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            str(15): nn.ReLU(inplace=True),
            str(16): nn.MaxPool2d(kernel_size=2, stride=2, padding=0, return_indices=True, dilation=1, ceil_mode=False),
            str(17): nn.Conv2d(256, 512, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            str(18): nn.ReLU(inplace=True),
            str(19): nn.Conv2d(512, 512, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            str(20): nn.ReLU(inplace=True),
            str(21): nn.Conv2d(512, 512, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            str(22): nn.ReLU(inplace=True),
            str(23): nn.MaxPool2d(kernel_size=2, stride=2, padding=0, return_indices=True, dilation=1, ceil_mode=False),
            str(24): nn.Conv2d(512, 512, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            str(25): nn.ReLU(inplace=True),
            str(26): nn.Conv2d(512, 512, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            str(27): nn.ReLU(inplace=True),
            str(28): nn.Conv2d(512, 512, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            str(29): nn.ReLU(inplace=True),
            str(30): nn.MaxPool2d(kernel_size=2, stride=2, padding=0, return_indices=True, dilation=1, ceil_mode=False)
        }))

    def forward(self, x, layer):
        indices = []
        print ("Input shape : ", str(x.shape))
        
#         for layer in range(layer):
#             if isinstance(self.features[layer], nn.MaxPool2d):
#                 print(x.shape)
#                 x, idx = layer(x)
#                 indices.append(idx)
            
            
        for layer in (self.features)[:layer]:
            if isinstance(layer, nn.MaxPool2d):
                x, idx = layer(x)
                print("forw pool : ", x.shape)
                
                indices.append(idx)
            else:
                x = layer(x)
                print("forw : ", x.shape)
                
                
        return x, indices
    
    def prints(self):
        print(len(self.features))
        print (type(self.features))



    
```

```python

class CNNLayerVisualization():
    """
        Produces an image that minimizes the loss of a convolution
        operation for a specific layer and filter
    """
    def __init__(self, model, selected_layer, selected_filter):
        self.model = model
        self.model.eval()
        self.selected_layer = selected_layer
        self.selected_filter = selected_filter
        self.conv_output = 0
        
       
        
        # Create the folder to export images if not exists
        if not os.path.exists('./generated'):
            os.makedirs('./generated')

    def hook_layer(self):
        def hook_function(module, grad_in, grad_out):
            # Gets the conv output of the selected filter (from selected layer)
            self.conv_output = grad_out[0: self.selected_filter]
        # Hook the selected layer
        self.model[self.selected_layer].register_forward_hook(hook_function)

    def visualise_layer_with_hooks(self):
        # Hook the selected layer
        self.hook_layer()
        # Generate a random image
        random_image = np.transpose(image[0])
        # Process image and return variable
#         processed_image = preprocess_image(random_image, False)
        processed_image = image
        # Define optimizer for the image
        optimizer = Adam([processed_image], lr=0.1, weight_decay=1e-6)
        for i in range(1, 31):
            optimizer.zero_grad()
            # Assign create image to a variable to move forward in the model
            x = processed_image
            indices = []
            for index, layer in enumerate(self.model):
                # Forward pass layer by layer
                # x is not used after this point because it is only needed to trigger
                # the forward hook function
                if isinstance(layer, nn.MaxPool2d):
#                     print (layer)
#                     print (x.shape)
                    x, idx = layer(x)
#                     print (idx.shape)
                    indices.append(idx)
#                     print(len(indices))
                else:
                    x = layer(x)
#                 x = layer(x)
#                 print (x.shape)
                # Only need to forward until the selected layer is reached
                if index == self.selected_layer:
                    # (forward hook function triggered)
                    break
            # Loss function is the mean of the output of the selected layer/filter
            # We try to minimize the mean of the output of that specific filter
#             print ("0 : ", self.conv_output[0])
#             print ("1 : ", self.conv_output[1])
#             de = deconv(self.conv_output, self.selected_layer, indices)
            print (len(indices))

            de = deconv(x, self.selected_layer, indices)
            print(len(self.conv_output))
            # print(x.shape)
            loss = -torch.mean(de)
#             loss = -torch.mean(de)
            print('Iteration:', str(i), 'Loss:', "{0:.2f}".format(loss.data.numpy()))
            # Backward
            loss.backward()
            # Update image
            optimizer.step()
            # Recreate image
#             plt.imshow(de)
            self.created_image = recreate_image(de)
            # Save image
            if i % 5 == 0:
                im_path = './generated/layer_vis_l' + str(self.selected_layer) + \
                    '_f' + str(self.selected_filter) + '_iter' + str(i) + '.jpg'
                save_image(self.created_image, im_path)

    def visualise_layer_without_hooks(self):
        # Process image and return variable
        # Generate a random image
        # random_image = np.uint8(np.random.uniform(150, 180, (224, 224, 3)))
        random_image = np.transpose(image[0])
        # Process image and return variable
        processed_image = preprocess_image(random_image, False)
        # Define optimizer for the image
        optimizer = Adam([processed_image], lr=12, weight_decay=1e-4)
        for i in range(1, 31):
            optimizer.zero_grad()
            # Assign create image to a variable to move forward in the model
            x = processed_image
            indices = []
            
            for index, layer in enumerate(self.model):
                # Forward pass layer by layer
                if isinstance(layer, nn.MaxPool2d):
                    print (layer)
                    print (x.shape)
                    x, idx = layer(x)
#                     print (idx.shape)
                    indices.append(idx)
#                     print(len(indices))
                else:
                    print ("NNNNNNN")
                    x = layer(x)
                if index == self.selected_layer:
                    # Only need to forward until the selected layer is reached
                    # Now, x is the output of the selected layer
                    break
            # Here, we get the specific filter from the output of the convolution operation
            # x is a tensor of shape 1x512x28x28.(For layer 17)
            # So there are 512 unique filter outputs
            # Following line selects a filter from 512 filters so self.conv_output will become
            # a tensor of shape 28x28
#             print(x.shape) 
#             de = deconv(x[0])
#             plt.imshow(de)
            de = deconv(x, 31, indices)
            
            self.conv_output = x[0, self.selected_filter]
            # Loss function is the mean of the output of the selected layer/filter
            # We try to minimize the mean of the output of that specific filter
#             loss = -torch.mean(self.conv_output)
            loss = -torch.mean(de)
            print('Iteration:', str(i), 'Loss:', "{0:.2f}".format(loss.data.numpy()))
            # Backward
            loss.backward()
            # Update image
            optimizer.step()
            # Recreate image
            self.created_image = recreate_image(de)
            # Save image
            if i % 5 == 0:
                im_path = './generated/layer_vis_l' + str(self.selected_layer) + \
                    '_f' + str(self.selected_filter) + '_iter' + str(i) + '.jpg'
                save_image(self.created_image, im_path)
```
